```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
pickling_on = open("algo_comp_N_10_K_6_d_3_T_100_mc_25","rb") # normal(large) kappa and low kappa for given param
```

```python
cumulated_regret_cb_mnl2 = pickle.load(pickling_on)
cumulated_regret_ucb_mnl = pickle.load(pickling_on)
cumulated_regret_ts_mnl = pickle.load(pickling_on)
cumulated_regret_ucb_mnl_klow = pickle.load(pickling_on)
cumulated_regret_ts_mnl_klow = pickle.load(pickling_on)
```

```python
pickling_on.close()
```

```python
plottable_cumulated_regret_cb_mnl2 = np.mean(cumulated_regret_cb_mnl2,axis=0)
plottable_cumulated_regret_ucb_mnl = np.mean(cumulated_regret_ucb_mnl,axis=0)
plottable_cumulated_regret_ts_mnl = np.mean(cumulated_regret_ts_mnl,axis=0)

plottable_cumulated_regret_ucb_mnl_klow = np.mean(cumulated_regret_ucb_mnl_klow,axis=0)
plottable_cumulated_regret_ts_mnl_klow = np.mean(cumulated_regret_ts_mnl_klow,axis=0)



plottable_cumulated_regret_cb_mnl2_std = np.std(cumulated_regret_cb_mnl2,axis=0)
plottable_cumulated_regret_ucb_mnl_std = np.std(cumulated_regret_ucb_mnl,axis=0)
plottable_cumulated_regret_ts_mnl_std = np.std(cumulated_regret_ts_mnl,axis=0)

plottable_cumulated_regret_ucb_mnl_klow_std = np.std(cumulated_regret_ucb_mnl_klow,axis=0)
plottable_cumulated_regret_ts_mnl_klow_std = np.std(cumulated_regret_ts_mnl_klow,axis=0)
```

```python
x=np.arange(1,100+1)
plt.style.use('seaborn-white')
# plt.rcParams['font.family'] 	= 'serif'
# plt.rcParams['font.serif'] 		= 'Ubuntu'
# plt.rcParams['font.monospace'] 	= 'Ubuntu Mono'
plt.rcParams['font.size'] 		= 15
plt.rcParams['axes.labelsize'] 	= 20
plt.rcParams['axes.titlesize'] 	= 30
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 15
plt.rcParams['figure.titlesize']= 30
plt.xlabel("Time Horizon")
plt.ylabel("Cumulative Regret")
plt.errorbar(x,np.array(plottable_cumulated_regret_cb_mnl2),yerr=np.array(plottable_cumulated_regret_cb_mnl2_std),ecolor='b',label='CB-MNL',color='b',alpha=.8)
plt.errorbar(x,plottable_cumulated_regret_ucb_mnl,yerr=plottable_cumulated_regret_ucb_mnl_std,ecolor='g',label='UCB-MNL',color='g',alpha=.8)
plt.errorbar(x,plottable_cumulated_regret_ts_mnl,yerr=plottable_cumulated_regret_ts_mnl_std,ecolor='r',label='TS-MNL',color='r',alpha=.8)
plt.legend()
plt.show()
```
